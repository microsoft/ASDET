## Initiializing MSTICPy

```python
# Core MSTICPy initialization for Notebooks
from msticpy.nbtools import nbinit
nbinit.init_notebook(namespace=globals());

# Load query providers (typically you'll be using just one)
qry_prov = QueryProvider("AzureSentinel")
```

## Connect To Workplace

```python
qry_prov.connect(WorkspaceConfig())
```

```python
tables = qry_prov.schema_tables
```

```python
import ipywidgets as widgets
from utils import cleanEntropy, cleanDuplicates, cleanRegexes, binarizeTable, cleanInvariant, cleanTable

dropdown = widgets.Dropdown(
    options=tables,
    description='Table:',
    value='OfficeActivity'
)

display(dropdown)
```

```python
# Get the most recent n rows

nRows = 10000
table = dropdown.value
queryString = f"{table} | take {nRows}"
df = qry_prov.exec_query(queryString)
```

## Cleaning Duplicate Columns

```python
result = cleanDuplicates(df)
result.head()
```

## Cleaning invariant columns

```python
result = cleanInvariant(df)
result.head()
```

## Cleaning Based on Entropy Threshold

```python
# Default values are 0-0.5
result = cleanEntropy(df, entropyLowerThreshold=0, entropyHigherThreshold=0.3)
result.head()
```

```python
## Cleaning Based on Regular Expressions
```

```python
# Removing Anything That contains the word time
result = cleanRegexes(df, regexes=['^.*Time.*$'])
result.head()
```

## Binarizes a Table Based on Presence/Absence

```python
result= binarizeTable(df)
result.head()
```

## General Cleaning Function

```python
result = cleanTable(df)
print(f"\nBefore: {df.shape[0]} x {df.shape[1]}\n")
display(df.head())
print(f"After: {result.shape[0]} x {result.shape[1]}")
display(result.head())
```
