# Time Series Anomaly Detection

```python
import pandas as pd

df = pd.read_csv('./data/occupancy/datatraining.csv')
df = df.drop(['index', 'date', 'Occupancy'], axis=1)
df.head()
```

## Simple Sliding Window Approach - Z-score

```python
window = 10
stdThreshold = 3
# zScoreDict = {col: [] for col in list(df)}
# for i in range(0, df.shape[0], window):
#     for key in zScoreDict.keys():
#         print(key, df[key].iloc[i:i+window])
```

```python
mean = df.rolling(window).mean()
std = df.rolling(window).std()
upperBound = mean+(3*std)
lowerBound = mean-(3*std)
withinLower = lowerBound.iloc[window:] <= df.iloc[window:]
withinUpper = df.iloc[window:] <= upperBound.iloc[window:]
```

```python
for rowIndex in range(window, df.iloc[window:].shape[0]):
    rows = []
    if not (withinUpper.iloc[rowIndex][0] == True and withinUpper.iloc[rowIndex].all()):
        print(f"Row {rowIndex} has some values that are more than {stdThreshold} std away")
        print(df.iloc[rowIndex])
    elif not (withinLower.iloc[rowIndex][0] == True and withinLower.iloc[rowIndex].all()):
        print(f"Row {rowIndex} has some values that are less than {stdThreshold} away")
        print(df.iloc[rowIndex])
```
